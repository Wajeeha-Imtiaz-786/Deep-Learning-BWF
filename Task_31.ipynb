```python
import numpy as np
from keras.utils import pad_sequences
from keras.preprocessing.text import Tokenizer
#from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense

# Example data
texts = ['I am a coder',
         'Harry Potter movies are interesting',
         'Mars is my favorite chocolate']

# One-hot encoding
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
one_hot_data = tokenizer.sequences_to_matrix(sequences, mode='binary')

# Word embeddings
max_sequence_length = max([len(seq) for seq in sequences])
word_index = tokenizer.word_index
num_words = len(word_index) + 1

embedding_dim = 100  # dimensionality of the word embeddings

# Pad sequences to have the same length
padded_sequences = pad_sequences(sequences, maxlen=max_sequence_length)

# Define the model
model = Sequential()
model.add(Embedding(num_words, embedding_dim, input_length=max_sequence_length))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
labels = np.array([1, 0, 1])  # example binary labels
model.fit(padded_sequences, labels, epochs=10, batch_size=1)

# Get the word embeddings
embedding_matrix = model.get_weights()[0]

# Print the word embeddings
for word, i in word_index.items():
    if i < num_words:
        print(word, embedding_matrix[i])

```
